# Una variable



## Histograma: Una variable continua

Los histogramas representan en altura la acumulación de casos en intervalos de valor. Una sola variable continua puede representarse de esta manera.

```
plt.hist(data_gaussian_1, bins=10);
plt.show()
```



Para el ejemplo, crearemos varias distribuciones de datos, y observaremos como se muestran en histogramas.

```
data_gaussian_2 = np.random.normal(loc=1.5, scale=0.1, size=500)
data_gaussian_3 = np.random.normal(loc=0.5, scale=0.6, size=500)

# Set an alpha and use the same bins since we are plotting two hists
plt.hist(data_gaussian_2, bins=20, alpha=0.5, label='data 1')
plt.hist(data_gaussian_3, bins=20, alpha=0.5, label='data 2')
plt.hist(data_gaussian_3, bins=20, alpha=0.5, label='data 3')

plt.legend(loc='best');

plt.show()
```



Pero se ve extraño, ¿porqué pasa?

```
bins = np.linspace(-3, 3, 50)

# Set an alpha and use the same bins since we are plotting two hists
plt.hist(data_gaussian_2, bins=bins, alpha=0.5, label='data 1')
plt.hist(data_gaussian_3, bins=bins, alpha=0.5, label='data 2')

plt.legend(loc='best');
plt.show()
```



## Barras: Una variable categórica

Para representar una variable categórica usamos el clásico diagrama de barras. Se representa en alturas (o longitudes) ciertas cantidades asociadas a ciertos valores.



```
cities = [
"New York",
"Chicago",
"San Francisco",
"Los Angeles",
"Montgomery"]

numbers = [np.random.randint(10) for city in cities]
positions = list(range(len(cities)))
print(cities)
print(positions)
print(numbers)
```

```
plt.bar(positions, numbers, align="center",tick_label=cities)
plt.show()
```

```
data_gaussian_4 = np.random.normal(loc=0.5, scale=2, size=500)
plt.scatter(data_gaussian_2, data_gaussian_3, alpha=0.5, marker="*");
plt.scatter(data_gaussian_2, data_gaussian_4, alpha=0.2, marker="8");


plt.show()
```



## Gráfica sencilla (Plot)

También se puede unir la secuencia de puntos, con lineas mediante un plot

```
x = np.linspace(0, 10, 10)
y = x ** 2

plt.plot(x, y)
plt.show()
```



A la que se le pueden añadir otros elementos: Varias lineas, leyenda, valores ... 

```
x = np.linspace(0, 2, 10)

plt.plot(x, x, 'o-', label='Linear') # modify type of representation and assigned label
plt.plot(x, x ** 2, 'x-', label='Quadratic')

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Linear and quadratic relationship')  # main title

plt.xlabel('Independent axis')
plt.ylabel('Dependent axis');

plt.show()
```
