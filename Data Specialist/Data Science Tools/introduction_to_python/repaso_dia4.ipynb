Crear diccionarios a partir de listas

```python
nombres = ['maria', 'jose', 'luis', 'mario']
edades = [12,34,56,78]
```

```python
clientes ={}
for key, value in zip(nombres,edades):
    clientes[key] = value
```

```python
list(zip(nombres,edades))
```

### List comprehension

```python
otralista = [1,2,3,4,5,20,40,55]
```

```python
l = []
for i in otralista:
    if i >= 10:
        l.append(i)
print(l)        
```

```python
[i for i in otralista if i >= 10]
```

```python
l = []
for i in otralista:
    if i >= 10:
        l.append(i)
    else:
        l.append('hola')
print(l)  
```

```python
[i if i >= 10 else 'hola' for i in otralista]
```
