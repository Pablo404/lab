## Importar módulos y paquetes

Python es un lenguaje de propósito general, por lo tanto si queremos utilizar estructuras de datos, funciones, y métodos adicionales, como: funciones matemáticas, estadísticas,o de procesamiento de texto, es necesario importar los módulos correspondientes para poder utlizarlos. Para aplicaciones cientificas con Python, algunas de las librerías más usadas son: numpy, matplotlib, scikit-learn, entre otros.

Existen tres formas diferentes para importar módulos de Python:

- Directamente importando toda la librería:

```python
import numpy
numpy.sum([2,3])
```
- Usando un alias (para poder llamar despues alguna función de la librería usando el alias):

```python
import numpy as np
np.sum([3,2])
```

- Importando sólo algunos módulos

```python
from numpy import sum
sum([3,4])
```



Una vez hecho eso, generalicelo metiéndolo en funciones, tal que se puedan pasar por argumento los números a calcularle el logaritmo, raíz etc. Pueden ser distintos.

````
# Respuesta

def arithmetic_func(log, log10, raiz):
    print("log("+str(log)+") =", math.log(log))
    print("log10("+str(log10)+", 10) =", math.log(log10, 10))
    print("sqrt("+str(raiz)+") =", math.sqrt(raiz)) 

def trigonometric_func(sin, cos, tan, deg, rad):
    print("sin("+str(sin)+") =", math.sin(sin))
    print("cos("+str(cos)+") =", math.cos(cos))
    print("tan("+str(tan)+") =", math.tan(tan))
    print("degrees("+str(deg)+") =", math.degrees(deg))
    print("radians("+str(rad)+") =", math.radians(rad))
    
args = (math.e, 10, 4.0)
print (arithmetic_func(*args))

args = (math.pi / 2, math.pi / 2, math.pi / 2, math.pi / 2, 90)
print (trigonometric_func(*args))
````

````
# Respuesta

import random
import time

def fast_quiz(numer_of_questions=5):
    correctCount = 0 # Count the number of correct answers
    count = 0 # Count the number of questions

    startTime = time.time() # Get start time

    while count < numer_of_questions:
        # 1. Generate two random single-digit integers
        number1 = random.randint(0, 9)
        number2 = random.randint(0, 9)

        # 2. If number1 < number2, swap number1 with number2
        if number1 < number2:
            number1, number2 = number2, number1

        # 3. Prompt the student to answer "what is number1 - number2?" 
        answer = int(input("What is " + str(number1) + " - " + 
            str(number2) + "? "))

        # 5. Grade the answer and display the result
        if number1 - number2 == answer:
            print("Correct!")
            correctCount += 1
        else:
            print("The wrong answer\n", number1, "-",
                number2, "it should be", (number1 - number2))

        # Increase the count
        count += 1

    endTime = time.time() # Get end time
    testTime = int(endTime - startTime) # Get test time
    return correctCount, testTime

number_of_questions = 5 

correctCount, testTime = fast_quiz()
print("The number of correct answers is", correctCount, "of", 
    number_of_questions, "\nThe time of the test has been", testTime, "seconds")
````
