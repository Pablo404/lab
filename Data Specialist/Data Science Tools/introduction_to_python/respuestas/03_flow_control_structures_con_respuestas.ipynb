## Estructuras de decisión y control



### Sentencias condicionales



En Python existen las sentencias condicionales `if`, `else`, `elif` (if else).

```python
if [condicion 1]:
    ...
elif [condicion 2]:
    ...
else:
    ...
```

````
age = 30
if age >= 18:
    print("I am inside the if")
    print("still inside the if")
    print("still inside the if")
print('I am outside the if')
````

````
age = 34

if age >= 18:
    print("Over 18")
    print("I am outside the if")
elif age >= 0 and age < 18:
    print("Under 18")
else:
    print("Not a valid age number")
````



La condición `if-then-else` también se puede escribir como una operación ternaria en una linea:

```python

```

### Ejercicio 1

Escribe código que solicite por pantalla la longitud de los lados del triángulo, y luego imprima si es equilatero, escaleno o isósceles. 

````
# Respuesta

print("Enter the length of the sides of the triangle: ")
x = int(input("x: "))
y = int(input("y: "))
z = int(input("z: "))

if x == y == z:
    print("Equilateral triangle")
elif x != y != z:
    print("Scalene Triangle")
else:
    print("Isosceles triangle") 
````

Escribe un programa que compruebe si un string representa un entero o no. Para ello, solicitará una cadena por pantalla, una vez introducida comprobará si representa un entero o no, y luego imprimirá la conclusión por pantalla.

Hint: lee sobre el método strip()

Ejemplo: 
- "163176391283". La cadena es un entero.
- "hsfusd12313hiho". La cadena no es un entero.
- "-4654". La cadena es un entero.


````
# Respuesta

text = input("Enter a string: ")
text = text.strip()
if len(text) < 1:
    print("Enter a valid text")
else:
    if all(text[i] in "0123456789" for i in range(len(text))):
        print("The string is a whole number.")
    elif (text[0] in "+-") and \
         all(text[i] in "0123456789" for i in range(1,len(text))):
        print("The string is a whole number.")
    else:
        print("The string is not a whole number.") 
````
