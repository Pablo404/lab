Slicing and indexing

```python
a = np.arange(64).reshape(4,4,4)
a
```

```python
a[1,2,1]
```

```python
a[1,:,2]
```

```python
a = np.arange(25).reshape(5,5)
a
```

```python
a[1,1] #fila 1 elemento 1
a[2,2] #fila 2 elemento 2
a[[1,2],[1,2]] #y así lo devolvemos todo completo
```
